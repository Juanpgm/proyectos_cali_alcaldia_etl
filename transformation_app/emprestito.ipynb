## CREACIÓN DE FOUNDATIONAL DIMS TABLE

```python
# Mantener solo las columnas mencionadas en columns_names_changes
columns_to_keep = list(columns_names_changes.values())
df_foundational_dims = df_foundational[columns_to_keep]
df_foundational_dims.head()
```

```python
# Reemplazar todos los valores NaN en df_foundational_dims con None o un valor específico
df_foundational_dims = df_foundational_dims.fillna('')  # Reemplaza NaN con cadenas vacías
# O alternativamente:
# df_foundational_dims = df_foundational_dims.fillna(0)  # Reemplaza NaN con 0
# df_foundational_dims = df_foundational_dims.where(pd.notnull(df_foundational_dims), None)  # Reemplaza NaN con None
```

```python
# Exportar df_foundational_dims como JSON
output_path_dims = 'app_outputs/emprestito_output/foundational_dims.json'

# Convertir el DataFrame a un diccionario y luego a JSON
df_foundational_dims_dict = df_foundational_dims.to_dict(orient='records')

# Limpiar el formato BPIN para eliminar el ".0"
for record in df_foundational_dims_dict:
    if 'bpin' in record and pd.notna(record['bpin']) and record['bpin'] != '':
        record['bpin'] = str(int(record['bpin']))

# Filtrar para mantener solo el periodo más reciente para cada BPIN
df_foundational_dims_latest = df_foundational_dims.copy()

# Crear una columna temporal para ordenar por fechas
df_foundational_dims_latest['fecha_temp'] = pd.to_datetime(df_foundational_dims_latest['fecha_terminacion'], errors='coerce')

# Ordenar por BPIN y fecha de terminación (más reciente primero)
df_foundational_dims_latest = df_foundational_dims_latest.sort_values(['bpin', 'fecha_temp'], ascending=[True, False])

# Mantener solo el primer registro (más reciente) para cada BPIN
df_foundational_dims_latest = df_foundational_dims_latest.groupby('bpin').first().reset_index()

# Eliminar la columna temporal
df_foundational_dims_latest = df_foundational_dims_latest.drop('fecha_temp', axis=1)

# Convertir a diccionario para exportar
df_foundational_dims_dict = df_foundational_dims_latest.to_dict(orient='records')

# Limpiar el formato BPIN para eliminar el ".0"
for record in df_foundational_dims_dict:
    if 'bpin' in record and pd.notna(record['bpin']) and record['bpin'] != '':
        record['bpin'] = str(int(record['bpin']))
with open(output_path_dims, 'w', encoding='utf-8') as output_file:
    json.dump(df_foundational_dims_dict, output_file, indent=2, ensure_ascii=False, default=str)

print(f"JSON guardado en: {output_path_dims}")
print(f"Número de registros exportados: {len(df_foundational_dims_dict)}")
```

## CREACIÓN DE FOUNDATIONAL FACTS TABLE

```python
columns_to_delete = [
    'centro_gestor',
    'banco',
    'bp',
    'descripcion_bp',
    'nombre_comercial',
    'proyectos_contratos',
    'valor_contrato',
    'tipo_contratacion',
    'pliego_tipo',
    'vig_futura',
    'deleg',
    'fecha_inicio',
    'fecha_terminacion',
    'cod_contrato',
    'cdp',
    'rpc',
    'link_secop',
    'fecha_publicacion_proceso',
    'fecha_adjudicacion',
    'observaciones',
    # Agrega más cambios de nombres según sea necesario
]
```

```python
# Crear df_foundational_facts como una copia de df_foundational
df_foundational_facts = df_foundational.copy()

# Eliminar las columnas especificadas en columns_to_delete
df_foundational_facts = df_foundational_facts.drop(columns=columns_to_delete)

# Mostrar las primeras filas del DataFrame resultante
df_foundational_facts.head()
```

```python
# Identificar las columnas que contienen "Unnamed", "Subtotal" y "GRAN TOTAL"
columns_to_drop = [col for col in df_foundational_facts.columns 
                  if 'Unnamed' in col or 'Subtotal' in col or 'GRAN TOTAL' in col]

# Eliminar las columnas identificadas
df_foundational_facts = df_foundational_facts.drop(columns=columns_to_drop)

# Mostrar las primeras filas del DataFrame actualizado
df_foundational_facts.head()
```

```python
df_foundational_facts.columns = df_foundational_facts.columns.str.lower()
```

```python
# Agrupar los datos por 'bpin' y sumar todas las columnas numéricas
df_foundational_facts_grouped = df_foundational_facts.groupby('bpin').sum(numeric_only=True).reset_index()

# Mostrar las primeras filas del DataFrame agrupado
df_foundational_facts_grouped.head()
```

```python
# Crear la estructura de datos anidada para el JSON
nested_data = {}

# Iterar sobre cada fila del DataFrame
for _, row in df_foundational_facts_grouped.iterrows():
    bpin = row['bpin']
    
    # Saltar filas con BPIN nulo o no válido
    if pd.isna(bpin):
        continue
    
    # Convertir BPIN a string y eliminar el ".0"
    bpin_str = str(int(bpin))
    
    # Inicializar el diccionario para este BPIN si no existe
    if bpin_str not in nested_data:
        nested_data[bpin_str] = {}
    
    # Obtener todas las columnas del DataFrame (excepto 'bpin')
    columnas = [col for col in df_foundational_facts_grouped.columns if col != 'bpin']
    
    for columna in columnas:
        # Saltar columnas de subtotales
        if 'subtotal' in columna.lower() or 'gran total' in columna.lower():
            continue
            
        # Parsear el nombre de la columna para extraer fecha y tipo
        partes = columna.split(' ')
        if len(partes) < 2:
            continue
            
        fecha_parte = partes[0]  # ej: "jul-25"
        tipo_parte = ' '.join(partes[1:]).lower()  # ej: "desembolso", "avance", "desembolso real", "avance real"
        
        # Extraer año y mes de la fecha
        if '-' in fecha_parte:
            month, year = fecha_parte.split('-')
            year = f"20{year}"  # Convertir "25" a "2025"
            # Mapear los nombres de meses en español a sus valores numéricos
            spanish_months = {
                'ene': '01', 'feb': '02', 'mar': '03', 'abr': '04', 'may': '05', 'jun': '06',
                'jul': '07', 'ago': '08', 'sep': '09', 'oct': '10', 'nov': '11', 'dic': '12'
            }
            
            # Convertir el mes a formato numérico (ISO)
            if month.lower() in spanish_months:
                month = spanish_months[month.lower()]
            else:
                raise ValueError(f"Mes no reconocido: {month}")
            fecha_iso = f"{year}-{month}"  # Formato ISO: YYYY-MM
        else:
            continue
        
        # Inicializar el diccionario para esta fecha si no existe
        if fecha_iso not in nested_data[bpin_str]:
            nested_data[bpin_str][fecha_iso] = {
                'desembolso': None,
                'avance': None,
                'desembolso_real': None,
                'avance_real': None
            }
        
        # Obtener el valor de la celda
        valor = row[columna] if not pd.isna(row[columna]) else None
        
        # Asignar el valor al tipo correspondiente
        if tipo_parte == 'desembolso':
            nested_data[bpin_str][fecha_iso]['desembolso'] = valor
        elif tipo_parte == 'avance':
            nested_data[bpin_str][fecha_iso]['avance'] = valor
        elif tipo_parte == 'desembolso real':
            nested_data[bpin_str][fecha_iso]['desembolso_real'] = valor
        elif tipo_parte == 'avance real':
            nested_data[bpin_str][fecha_iso]['avance_real'] = valor

# Convertir a JSON y guardar
output_path = 'app_outputs/emprestito_output/foundational_facts.json'
with open(output_path, 'w', encoding='utf-8') as output_file:
    json.dump(nested_data, output_file, indent=2, ensure_ascii=False, default=str)

print(f"JSON guardado en: {output_path}")
print(f"Número de BPINs procesados: {len(nested_data)}")
```
